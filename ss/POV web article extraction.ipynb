# POV web article scraper

```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
def scrape_text_from_url(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract text from the div with ID "article-title"
        article_title = soup.find('div', {'id': 'article-title'})
        article_title_text = article_title.get_text() if article_title else ''

        # Extract text from the div with class "content-components"
        content_components = soup.find('div', {'class': 'content-components'})
        content_components_text = content_components.get_text() if content_components else ''

        return {
            'article_title': article_title_text,
            'content_components': content_components_text
        }
    else:
        # Print an error message if the request was unsuccessful
        print(f"Failed to retrieve content. Status code: {response.status_code}")
        return None

# Drop url to article you want to scrape here
url_to_scrape = 'https://www2.deloitte.com/ca/en/pages/public-sector/articles/emerging-tech-smart-cities-safer.html'
result = scrape_text_from_url(url_to_scrape)

if result:
    print(f"Article Title: {result['article_title']}")
    print(f"Content Components: {result['content_components']}")
else:
    print("Scraping failed.")
```
